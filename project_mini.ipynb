# Handling Outliers in data

```python
#importing matplot library for data visualization
import matplotlib.pyplot as plt
```

```python
#code for box plot for all columns
c=0
plt.figure(figsize=(18, 10))
for i in df.drop(columns=['Warehouse_block', 'Mode_of_Shipment','Product_importance', 'Gender', 'Reached.on.Time_Y.N','ID']).columns:
  if str(df[i].dtype)=='object':
    continue
  plt.subplot(2, 3, c+1)
  plt.boxplot(df[i])
  plt.title(i)
  c+=1
plt.show()
```

```python
#removing the outliers for 'PRIPR_PURCHASES' column
#calculating the Inter Quartile Range,q3,q1
IQR = df['Prior_purchases'].quantile(0.75)-df['Prior_purchases'].quantile(0.25)
lb=df['Prior_purchases'].quantile(0.25)-(1.5*IQR)
ub=df['Prior_purchases'].quantile(0.75)+(1.5*IQR)
df['Prior_purchases']=np.where(df['Prior_purchases']>ub,ub,df['Prior_purchases'])
sns.boxplot(df['Prior_purchases'])
```

```python
#removing outliers
IQR1 = df['Discount_offered'].quantile(0.75)-df['Discount_offered'].quantile(0.25)
lb1=df['Discount_offered'].quantile(0.25)-(1.5*IQR1)
ub1=df['Discount_offered'].quantile(0.75)+(1.5*IQR1)
df['Discount_offered']=np.where(df['Discount_offered']>ub1,ub1,df['Discount_offered'])
sns.boxplot(df['Discount_offered'])
```

```python
#code for box plot for some columns
c=0
plt.figure(figsize=(18, 10))
for i in df.drop(columns=['Customer_care_calls', 'Customer_rating','Cost_of_the_Product', 'Prior_purchases','Discount_offered','Weight_in_gms', 'Reached.on.Time_Y.N','ID']).columns:
  if str(df[i].dtype)=='object':
    continue
  plt.subplot(2, 3, c+1)
  plt.boxplot(df[i])
  plt.title(i)
  c+=1
plt.show()
```

# MODEL BUILDING

SPLITING THE DATA

```python
#spliting the data set in to x and y variables
#x is an independent variable
#y is an dependent variable
x=df.drop(columns=['ID','Reached.on.Time_Y.N'],axis=1)
y=df['Reached.on.Time_Y.N']
```

```python
x.head()
```

```python
y.head()
```

SCALING THE DATA

```python
#importing library for scaling the dataset
from sklearn.preprocessing import StandardScaler
```

```python
#initializing the scaler
#fit and transforming in to standard data
sc=StandardScaler()
scaled_x=sc.fit_transform(x)
scaled_x
```

```python
#array converting in to data frame
scaled_x=pd.DataFrame(x)
scaled_x
```

TRAIN,TEST,SPLIT

```python
#importing the library for train,test and split
from sklearn.model_selection import train_test_split
```

```python
#define the train,test
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

#TRAINING THE MODEL IN MULTIPLE ALGORITHMS

LOGISTIC REGRESSION

```python
#LOGISTIC REGRESSION
#importing the library
from sklearn.linear_model import LogisticRegression
#initializing the model
lr=LogisticRegression()
#fit the model
lr.fit(x_train,y_train)
#predict the model
predic=lr.predict(x_test)
#finding accuracy,classification report
from sklearn.metrics import classification_report
print(classification_report(y_test,predic))
```

DECISION TREE CLASSIFIER

```python
#decision tree classifier
from sklearn.tree import DecisionTreeClassifier
dec=DecisionTreeClassifier()
dec.fit(x_train,y_train)
predi=dec.predict(x_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,predi))
```

RANDOM FOREST CLASSIFIER

```python
#random forest classifier
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(x_train,y_train)
pred=rfc.predict(x_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,pred))
```

SUPPORT VECTOR MACHINE-LINEAR

```python
from sklearn import svm
lin=svm.SVC(kernel="linear")
lin.fit(x_train,y_train)
pre=lin.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(pre,y_test)
```

SUPPORT VECTOR MACHINE-NON-LINEAR

```python
from sklearn import svm
pol=svm.SVC(kernel="poly")
pol.fit(x_train,y_train)
pr=pol.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(pr,y_test)
```

K-NEAREST NEIGHBORS

```python
#knn
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
p=knn.predict(x_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,p))
```

XGBOOST CLASSIFIER

```python
#XGBOOST CLASSIFICATION
from xgboost import XGBClassifier
xg=XGBClassifier()
xg.fit(x_train,y_train)
p=xg.predict(x_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,p))
```

RIDGE CLASSIFIER

```python
#RIDGE CLASSIFIER
from sklearn.linear_model import RidgeClassifier
rg=RidgeClassifier()
rg.fit(x_train,y_train)
p=rg.predict(x_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,p))
```

TESTING THE MODEL

BEFORE HYPER PARAMETER TUNNING

```python
#importing the library for metrics
from sklearn.metrics import f1_score,recall_score,precision_score
```

```python
def eval(name,model):
  y_pred=model.predict(x_test)
  result=[]
  result.append(name)
  result.append("{:.2f}".format(accuracy_score(y_test,y_pred)*100))
  result.append("{:.2f}".format(f1_score(y_test,y_pred)*100))
  result.append("{:.2f}".format(recall_score(y_test,y_pred)*100))
  result.append("{:.2f}".format(precision_score(y_test,y_pred)*100))
  return result
model_list={
    'Logistic Regression':lr,
    'Decision Tree Classifier':dec,
    'Random Forest':rfc,
    'Support Vector Classifier-linear':lin,
    'Support Vector Classifier-non-linear':pol,
    'K-Nearest Nieghbors':knn,
    'Xgboost':xg,
    'Ridge Classifier':rg,
  }
model_eval_info=[]
for i in model_list.keys():
  model_eval_info.append(eval(i,model_list[i]))
model_eval_info=pd.DataFrame(model_eval_info,columns=['Name','Accuracy','f1_score','Recall','Precision'])
model_eval_info.to_csv('model_eval.csv')
model_eval_info
```

# Hyper Parameter Tunning

GRID SEARCH

```python
#importing the library for grid search
from sklearn.model_selection import GridSearchCV
```

LOGISTIC REGRESSION HYPER PARAMETER TUNNING

```python
#finding the grid search cv for logistic regression
lr=LogisticRegression(n_jobs=-1,random_state=0)
lr_param_grid={
    'C':[0.1,0.5,1,5,10],
    'solver':['liblinear','saga'],
    'penalty':['l1','l2']

}
lr_cv=GridSearchCV(lr,lr_param_grid,cv=5,scoring="accuracy",n_jobs=-1,verbose=True)
lr_cv.fit(x_train,y_train)
```

RANDOM FOREST HYPER PARAMETER TUNNING

```python
#finding the grid search cv for random forest classifier
rfc=RandomForestClassifier()
rfc_param_grid={
    'n_estimators':[100,200],
    'criterion':['entropy','gini'],
    'max_depth':[5,10],
    'max_features':['auto','sqrt']
}
rfc_cv=GridSearchCV(rfc,rfc_param_grid,cv=3,scoring="accuracy",n_jobs=-1,verbose=3)
rfc_cv.fit(x_train,y_train)
```

XGBOOST CLASSIFIER-HYPER PARAMETER TUNNIG

```python
#finding the grid search cv for xgboost
params={
    'min_child_weight':[10,20],
    'gamma':[1.5,2.0,2.5],
    'colsample_bytree':[0.6,0.8,0.9],
    'max_depth':[4,5,6]
}
xg=XGBClassifier(learning_rate=0.5,n_estimators=100,objective='binary:logistic',nthreads=3)
xg_cv=GridSearchCV(xg,param_grid=params,cv=5,refit=True,scoring="accuracy",n_jobs=-1,verbose=3)
xg_cv.fit(x_train,y_train)
```

DECISION TREE CLASSIFIER-HYPER PARAMETER TUNNING

```python
#finding grid search cv for decision tree classifier
dec=DecisionTreeClassifier(random_state=42)
params={ 'max_depth': [2, 3, 5, 10, 20],
    'min_samples_leaf': [5, 10, 20, 50, 100],
    'criterion': ['gini', 'entropy']
}
dec_cv=GridSearchCV(dec,param_grid=params,cv=5,n_jobs=-1,scoring="accuracy",verbose=3)
dec_cv.fit(x_train,y_train)
```

RIDGE-CLASSIFIER-HYPER PARAMETER TUNNING

```python
#finding the grid search cv for ridge classifier
rg=RidgeClassifier(random_state=42)
params={
    'alpha':(np.logspace(-8,8,100))
}
rg_cv=GridSearchCV(rg,param_grid=params,cv=5)
rg_cv.fit(x_train,y_train)
```

kNN-CLASSIFIER

```python
knn=KNeighborsClassifier()
params={
    'n_neighbors':[3,5,7,9,11],
    'weights':['uniform','distance'],
    'metric':['euclidean','manhattan']

}
knn_cv=GridSearchCV(knn,param_grid=params,cv=5)
knn_cv.fit(x_train,y_train)
```

AFTER HYPER PARAMETER TUNNING

```python
def eval(name,model):
  y_pred=model.predict(x_test)
  result=[]
  result.append(name)
  result.append("{:.2f}".format(accuracy_score(y_test,y_pred)*100))
  result.append("{:.2f}".format(f1_score(y_test,y_pred)*100))
  result.append("{:.2f}".format(recall_score(y_test,y_pred)*100))
  result.append("{:.2f}".format(precision_score(y_test,y_pred)*100))
  return result
model_list={
    'Logistic Regression':lr_cv,
    'Decision Tree Classifier':dec_cv,
    'Random Forest':rfc_cv,
    'k-Nearest Nieghbors':knn_cv,
    'Xgboost':xg_cv,
    'Ridge Classifier':rg_cv,
  }
model_eval_info=[]
for i in model_list.keys():
  model_eval_info.append(eval(i,model_list[i]))
model_eval_info=pd.DataFrame(model_eval_info,columns=['Name','Accuracy','f1_score','Recall','Precision'])
model_eval_info.to_csv('model_eval.csv')
model_eval_info
```
